```python
#strip text
country_df['vaccines'] = country_df['vaccines'].str.replace("/Beijing|/Wuhan|/HayatVax|/BioNTech|Oxford/", "", regex=True)

#replace with Chinese vaccins
country_df['vaccines'] = country_df['vaccines'].str.replace("Sinopharm|BBIBP-CorV|Sinovac|CanSino|RBD-Dimer", "Chinese", regex=True)

#replace with Russian vaccins
country_df['vaccines'] = country_df['vaccines'].str.replace("EpiVacCorona|Sputnik V", "Russian", regex=True)

#replace with Cuban vaccins
country_df['vaccines'] = country_df['vaccines'].str.replace("Soberana02|Abdala", "Cuban", regex=True)

#replace some others
to_replace = {'Covaxin': 'Indian',
              'QazVac': 'Kazachstan'}
country_df['vaccines'] = country_df['vaccines'].replace(to_replace, regex=True)
```

```python
# Get vaccinations list
vac_list = [x.split(", ") for x in country_df.vaccines.values]
vaccins = [item for elem in vac_list for item in elem]
vaccins = set(vaccins)
vaccins = list(vaccins)
vaccins
```

```python
for vaccin in vaccins:
    d = country_df[vaccin] = np.where(country_df['vaccines'].str.contains(vaccin), True, False)

country_df = country_df.sort_values(by = ['country', 'date'], ascending = [True, False])
country_vacc_corr = country_df.drop_duplicates(subset = "country", keep = "first")
country_vacc_corr.iloc[:, np.r_[0,12, 15:len(country_vacc_corr.columns)]].head()
```

```python
def vaccine_map(color, vaccin):
    fig = px.choropleth(country_vacc_corr, locations = "iso_code",
                        color = vaccin,
                        hover_name = "country",
                        color_discrete_map = {True: color, False: 'lightgrey'})

    layout = go.Layout(
        title = go.layout.Title(
            text = f"<b>Countries using {vaccin} vaccin</b>",
            x = 0.5
        ),
        showlegend = False,
        font   = dict(size = 14),
        width  = 750,
        height = 350,
        margin =dict(l = 0,r = 0,b = 0,t = 30)
    )

    fig.update_layout(layout)
    fig.write_image('/kaggle/working/1.webp')
    fig.show()
```

```python
#function as I will make a similar map later
def plot_ww_numbers(data, color, hover_data, title):
    fig = px.choropleth(data, locations="iso_code",
                        color= color,
                        hover_data= hover_data)

    layout = go.Layout(title=go.layout.Title(
        text= f"<b>{title}</b>",
        x=0.5
        ),
        font=dict(size=14),
        width = 750,
        height = 350,
        margin=dict(l=0,r=0,b=0,t=30)
                      )

    fig.update_layout(layout)
    
    fig.show()
    fig.write_html('/kaggle/working/total_vacc_pred_fig_html.html')
```

```python
plot_ww_numbers(data  = country_vacc_corr ,
                color = 'Vaccins_used',
                hover_data= ["country", "vaccines"],
                title = 'Number of different vaccines used by country')
```
